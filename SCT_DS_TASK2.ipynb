**Overview of Health Insurance Cost Analysis**

Health insurance cost analysis examines factors influencing premium rates, medical expenses, and overall affordability.
Key determinants include demographics, health conditions, plan types, and medical inflation.
By leveraging data analytics, trends in claims, risk assessment, and cost projections can be identified. 
Insights from this analysis help individuals, insurers, and policymakers optimize plans, reduce unnecessary costs, and improve financial sustainability in healthcare.




**Task 2 - Exploratory Data Analysis (EDA)**


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv('insurance.csv')
print("\nDataset Loaded....")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Statistical summary
print(df.describe())
```

```python
# Distribution of 'sex'
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', data=df)
plt.title('Gender Distribution')
plt.show()
```

```python
# Age distribution using KDE
plt.figure(figsize=(8, 6))
sns.kdeplot(df['age'], fill=True)  # Use fill=True to shade the area under the curve
plt.xlabel("Age")
plt.ylabel("Density")
plt.title('Age Distribution (KDE)')
plt.show()

```

```python
# Smoker distribution using Pie Chart
smoker_counts = df['smoker'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(smoker_counts, labels=smoker_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Smoker Distribution (Pie Chart)')
plt.show()
```

```python
# Distribution of 'region'
plt.figure(figsize=(8, 6))
sns.countplot(x='region', data=df, color='red')
plt.title('Region Distribution')
plt.show()
```

```python
# Relationship between 'smoker' and 'charges'
plt.figure(figsize=(8, 6))
sns.boxplot(x='smoker', y='charges', data=df)
plt.title('Charges by Smoker Status')
plt.show()
```

```python
# Relationship between 'sex' and 'charges'
plt.figure(figsize=(8, 6))
sns.boxplot(x='sex', y='charges', data=df)
plt.title('Charges by Gender')
plt.show()
```

```python
# Correlation matrix (for numerical features)
numerical_cols = ['age', 'bmi', 'children', 'charges'] # Select numerical columns
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Relationship between 'region' and 'charges'
plt.figure(figsize=(10, 6))  # Increased figure size for better readability
sns.boxplot(x='region', y='charges', data=df)
plt.title('Charges by Region')
plt.show()
```
